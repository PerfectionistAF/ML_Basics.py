# GENERATE THE DATASETS, X1, X2

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns;
sns.set()
```

```python
#Generate random dataset given mean, covariance, and sample size
X1_class1_mean = (-8, 8)
X1_class2_mean = (8, 8)
cov = [[0.3, 1.5],[1.5, 9.0]]
X1_class1 = np.random.multivariate_normal(X1_class1_mean, cov, size = 200)
X1_class2 = np.random.multivariate_normal(X1_class2_mean, cov, size = 200)

X2_class1_mean = (-1, 0)
X2_class2_mean = (1, 0)
X2_class1 = np.random.multivariate_normal(X2_class1_mean, cov, size = 200)
X2_class2 = np.random.multivariate_normal(X2_class2_mean, cov, size = 200)
```

```python
#Plot all datasets in a single graph
plt.scatter(X1_class1[:, 0], X1_class1[:, 1], alpha=0.5 , color='red')
plt.scatter(X1_class2[:, 0], X1_class2[:, 1], alpha=0.5 , color='blue')

plt.scatter(X2_class1[:, 0], X2_class1[:, 1], alpha=0.5 , color='orange')
plt.scatter(X2_class2[:, 0], X2_class2[:, 1], alpha=0.5 , color='green')

plt.axis('auto')
plt.grid()
plt.show()
```

```python
#Plot all datasets in a single graph
from sklearn.decomposition import PCA
plt.scatter(X1_class1[:, 0], X1_class1[:, 1], alpha=0.5 , color='red')
plt.scatter(X1_class2[:, 0], X1_class2[:, 1], alpha=0.5 , color='blue')

plt.axis('auto')
plt.grid()
plt.show()
```

```python
plt.scatter(X2_class1[:, 0], X2_class1[:, 1], alpha=0.5 , color='orange')
plt.scatter(X2_class2[:, 0], X2_class2[:, 1], alpha=0.5 , color='green')

plt.axis('auto')
plt.grid()
plt.show()
```
