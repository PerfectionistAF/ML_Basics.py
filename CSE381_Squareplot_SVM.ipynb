```python
# Make a prediction with weights
# Make a prediction with weights
def predict(row, weights):
    activation = weights[0]
    for i in range(2):  ##Predict among pairs of values
        activation += weights[i + 1] * row[i]
    return 1.0 if activation >= 0.0 else 0.0
```

```python
#Perceptron alg on array
from sklearn.model_selection import train_test_split
weights_i =np.array([1, 1, -0.5])
weights_i_trans=weights_i.transpose()
print(f'Transposed Array:\n{weights_i_trans}\n')

prediction = predict(Xi_flat, weights_i_trans)
for i in range(10): ###Analysis is on full 100, this is for illustration purposes
    print("--",Xi_flat[i],"--",prediction)
    #print("Expected=%d, Predicted=%d" % (predict(row, weights_i_trans)))
```

```python
# Estimate Perceptron weights using stochastic gradient descent
def train_weights(train, l_rate, n_epoch):
    weights = [0.0 for i in range(2)]
    for epoch in range(n_epoch):
        sum_error = 0.0
        for i_class in range(100):
            prediction = predict(i_class, weights)
            error = i_class[-1] - prediction
            sum_error += error**2
            weights[0] = weights[0] + l_rate * error
            for j in range(2):
                weights[j + 1] = weights[j + 1] + l_rate * error * i_class[j]
        print('>epoch=%d, lrate=%.3f, error=%.3f' % (epoch, l_rate, sum_error))
    return weights
```

#### Since (w1/w2 are weights, bias is the last value given in parameter vector)
#### activation = (w1 * X1) + (w2 * X2) + bias (X1 and X2 being datasets) 
#### The function of the bias is to increase or decrease model accuracy upon activation
#### As given in the above equation, a negative bias is said to decrease accuracy in a perceptron that generates +ve values
#### such as the one given in X3. A positive bias increases accuracy so Expected VS Predicted have closer penalties/costs.
